```python
import pandas as pd

posts16 = pd.read_csv('tier16_liwc.csv.gz')
```

```python
posts16[['post.id','blog.id','posemo','negemo']].tail(10)
```

```python
def polarity(post):
    return post['posemo'] - post['negemo']

posts16['polarity'] = posts16.apply(polarity, axis=1)
```

```python
posts16[['post.id','blog.id','posemo','negemo','polarity']].tail(10)
```

```python
files = pd.read_csv('fileEmotionProb.csv.gz', encoding='iso-8859-1', header=None)
files.head(5)
```

```python
images = pd.read_csv('images.log', header=None, sep=':', error_bad_lines=False,)
images.head(5)
```

```python
import sys

posts16['img_polarity'] = ''
posts16['img'] = ''

for idx_img in images.index:
    post = images.iloc[idx_img,0].split('/')
    img = images.iloc[idx_img,2]
    
    if len(post) != 4:
        continue
    
    ## Busca Sentimento da Imagem
    img_like = '/home/greice/projeto/tier-16-images/' + img.replace('//','')
    if len(files[files[0]==img_like]) == 1:
        img_polarity = files[files[0]==img_like][1].values[0]
        
    ## Atribui valor da imagem ao post
    idx_post = posts16[(posts16['blog.id']==int(post[1])) & (posts16['post.id']==int(post[3].replace('.json','')))].index
    if len(idx_post) == 1:
        posts16.loc[idx_post,'img_polarity'] = img_polarity
        posts16.loc[idx_post,'img'] = img.replace('//','')
        
    if (idx_img % 1000) == 0:
        sys.stdout.write(str(idx_img) + ',')
        sys.stdout.flush()
        
```
